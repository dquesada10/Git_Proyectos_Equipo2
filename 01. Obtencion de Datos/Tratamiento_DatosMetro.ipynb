```python
#importamos las librerias
import pandas as pd
import numpy as np
from rdflib import Graph, Literal, RDF, URIRef, Namespace #basic RDF handling
from rdflib.namespace import FOAF , XSD #most common namespaces
import urllib.parse #for parsing strings to URI'
```

```python
#importamos como dataframes los csvs
stops_m4=pd.read_csv('stops_m4_ini.txt')
stops_m10=pd.read_csv('stops_m10_ini.txt')
```

```python
#Incluimos las variables identificativas de cada dataset y las juntamos
stops_m4["transportmean_name"]="METRO"
stops_m10["transportmean_name"]="ML"
```

```python
stops_m4["stop_timezone"].unique()
stops_m10["stop_timezone"].unique()
```

```python
stops_m4["stop_timezone"]='Europe/Madrid'
stops_m10["stop_timezone"]='Europe/Madrid'
```

```python
stops_m4["parent_station"] = np.where(stops_m4["location_type"] == 1, stops_m4["stop_id"], stops_m4["parent_station"])
stops_m4.info()
stops_m4.head(20)
```

```python
stops_m10["parent_station"] = np.where(stops_m10["location_type"] == 1, stops_m10["stop_id"], stops_m10["parent_station"])
stops_m10.info()
stops_m10.head(50)
```

```python
stops_m4["col_aux"]="est"+  stops_m4[stops_m4["parent_station"].isnull()]["stop_id"].str.slice(start=3)
```

```python
stops_m10["col_aux"]="est"+  stops_m10[stops_m10["parent_station"].isnull()]["stop_id"].str.slice(start=3)
```

```python
stops_m4["parent_station"]=np.where(stops_m4["col_aux"].notna(), stops_m4["col_aux"],stops_m4["parent_station"])
```

```python
stops_m10["parent_station"]=np.where(stops_m10["col_aux"].notna(), stops_m10["col_aux"],stops_m10["parent_station"])
```

```python
stops_m4.head(50)
stops_m4.info()
stops_m4[stops_m4["parent_station"].isna()]
```

```python
stops_m10.head(50)
stops_m10.info()
stops_m10[stops_m10["parent_station"].isna()]
```

```python
stops_m4["zone_id"].unique()
```

```python
stops_m10["zone_id"].unique()
```

```python

stops_m4_parents= stops_m4[stops_m4["location_type"]==1][["zone_id","parent_station"]]
stops_m4_parents
```

```python

stops_m10_parents= stops_m10[stops_m10["location_type"]==1][["zone_id","parent_station"]]
stops_m10_parents
stops_m4
```

```python
stops_m4=stops_m4.merge(stops_m4_parents, on='parent_station', how='left')
stops_m4['zone_id'] = stops_m4['zone_id_x'].fillna(stops_m4['zone_id_y'])
stops_m4.info()
```

```python
cols=['stop_id', 'stop_code', 'stop_name', 'stop_desc', 'stop_lat','stop_lon', 'zone_id', 'stop_url', 'location_type', 'parent_station','stop_timezone', 'wheelchair_boarding']
stops_m4=stops_m4[cols]
stops_m4.info()
```

```python
stops_m10=stops_m10.merge(stops_m10_parents, on='parent_station', how='left')
stops_m10['zone_id'] = stops_m10['zone_id_x'].fillna(stops_m10['zone_id_y'])
stops_m10=stops_m10[cols]
stops_m10.info()
```

```python
typeST_dict = {0:'parada', 1:'estacion', 2:'acceso'}
stops_m4['location_new'] = stops_m4['location_type'].map(typeST_dict)
```

```python
stops_m10['location_new'] = stops_m10['location_type'].map(typeST_dict)
```

```python
stops_m4.to_csv("stops_m4.txt")
```

```python
stops_m10.to_csv("stops_m10.txt")
```
